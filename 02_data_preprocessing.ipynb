# Data Preprocessing

## Objective

The objective of this notebook is to clean and prepare the river water quality dataset for exploratory data analysis and visualization.

```python
#Import the required libraries for data preprocessing.

import pandas as pd
import numpy as np
```

```python
# Load Dataset

import requests

url = "https://raw.githubusercontent.com/ARizalma/river-water-quality-analysis/main/Data/river-water-quality-data-for-2023.csv"

df = pd.read_csv(url, sep=";")
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.rename(columns={
    'titik_sampel': 'sampling_point',
    'lintang_selatan': 'latitude',
    'bujur_timur': 'longitude',
    'jenis_parameter': 'parameter_category',
    'parameter': 'parameter_name',
    'baku_mutu': 'water_quality_standard',
    'hasil_pengukuran': 'measurement_value',
    'periode_data': 'year',
    'periode_pemantauan': 'monitoring_period',
    'bulan_sampling': 'sampling_month',
    'nama_sungai': 'river_name',
    'alamat': 'location'
})
```

```python
df.columns
```

```python
df[['measurement_value',
    'water_quality_standard',
    'latitude',
    'longitude']].sample(20, random_state=55)
```

## Clean Decimal Separator
Convert decimal separators from commas to periods for numerical processing.

```python
# Replace comma decimal separators with periods

df['measurement_value'] = (
    df['measurement_value']
    .astype(str)
    .str.replace(',', '.', regex=False)
)

df['latitude'] = (
    df['latitude']
    .astype(str)
    .str.replace(',', '.', regex=False)
)

df['longitude'] = (
    df['longitude']
    .astype(str)
    .str.replace(',', '.', regex=False)
)
```

```python
df[['measurement_value', 'latitude', 'longitude']].head()
```

##Convert Data Types
Convert cleaned numerical columns into numeric data types.

```python
# Convert columns to numeric data types

df['measurement_value'] = pd.to_numeric(
    df['measurement_value'],
    errors='coerce'
)

df['latitude'] = pd.to_numeric(
    df['latitude'],
    errors='coerce'
)

df['longitude'] = pd.to_numeric(
    df['longitude'],
    errors='coerce'
)
```

```python
df[['measurement_value',
    'latitude',
    'longitude']].dtypes
```

## Fix Inconsistent Categories

Standardize inconsistent category labels.

```python
# Standardize category labels

df['parameter_category'] = (
    df['parameter_category']
    .replace({
        'FIsika': 'Fisika'
    })
)
```

```python
df['parameter_category'].value_counts()
```

## Validate Cleaned Dataset

Validate the dataset after preprocessing.

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.sample(10, random_state=55)
```

## Save Clean Dataset

Save the cleaned dataset for further analysis.

```python
df.to_csv(
    'river-water-quality-data-clean.csv',
    index=False
)
```
